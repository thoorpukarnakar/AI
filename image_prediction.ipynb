```python
#Train a ConvNet on the image data, This Data consists of 10 labels

from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
from keras.preprocessing.image import load_img, img_to_array
import keras
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
Y_train = pd.read_csv(r'./input/train.csv')
Y_test = pd.read_csv(r'./input/test.csv')


print(Y_train.shape)
#print(Y_test)
Y_train = np.array(Y_train)[::,1]
#Y_test = np.array(Y_test)[::,1]
#print(Y_train[])
```

```python
for i in Y_test['id']:
    print(i)
```

```python
X_train = [np.array(load_img(r'./input/train/{}.png'.format(i),\
                             target_size=(28,28),grayscale=True))/255 for i in tqdm(range(1,60001))]
```

```python
X_test = [np.array(load_img(r'./input/test/{}.png'.format(i),\
                            target_size=(28,28),grayscale=True))/255 for i in tqdm(Y_test['id'])]
```

```python
X_train = np.array(X_train)
X_train.shape
```

```python
X_test = np.array(X_test)
X_test.shape
```

```python
img_rows, img_cols = 28,28
batch_size = 128
n_echos = 25
num_clasess = 10
```

```python
if K.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0],1,img_rows,img_cols)
    X_test = X_test.reshape(X_test.shape[0],1,img_rows,img_cols)
    input_shape = (1,img_rows,img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0],img_rows,img_cols,1)
    X_test = X_test.reshape(X_test.shape[0],img_rows,img_cols,1)
    input_shape = (img_rows,img_cols,1)

    
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
```

```python
Y_train = keras.utils.to_categorical(Y_train,num_clasess)
Y_train.shape
```

```python
# Validation data set
x_train, x_test, val_train,val_test = train_test_split(X_train,Y_train,test_size = 0.33,random_state = 45)
```

```python
print(X_train.shape)
print(X_test.shape)
print(val_train.shape)
print(val_test.shape)

```

```python
model = Sequential()
model.add(Conv2D(64,kernel_size=(3,3),
                activation = 'relu',
                input_shape = input_shape))
#model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_clasess,activation='softmax'))

model.compile(loss = keras.losses.categorical_crossentropy,
             optimizer=keras.optimizers.Adadelta(),
             metrics=['accuracy'])

    
```

```python
model.summary()
```

```python
best_weights_filepath=r"./Best_model/weights-improvement.hdf5"
saveBestModel = ModelCheckpoint(best_weights_filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
hist = model.fit(x_train,val_train,
                batch_size = batch_size,
                epochs = n_echos,
                verbose = 1,
                 callbacks= [saveBestModel],
                validation_data = (x_test,val_test))
```

```python
score = model.evaluate(x_test,val_test,verbose= 0)
print('Test_loss:',score[0])
print('Test_accuracy:',score[1])
```

```python
model.load_weights(best_weights_filepath)
score = model.evaluate(x_test,val_test,verbose= 0)
print('Test_loss:',score[0])
print('Test_accuracy:',score[1])
```
